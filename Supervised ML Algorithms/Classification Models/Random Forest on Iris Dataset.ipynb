```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
```

```python
# Load dataset
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
# Train Random Forest Model
rf_model = RandomForestClassifier(n_estimators=5, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Create Figure for Animation
fig, ax = plt.subplots(figsize=(10, 6))

def update(frame):
    ax.clear()
    plot_tree(rf_model.estimators_[frame], filled=True, feature_names=iris.feature_names, ax=ax)
    ax.set_title(f"Decision Tree {frame+1}")

# Generate Animation
ani = animation.FuncAnimation(fig, update, frames=len(rf_model.estimators_), repeat=True)
ani.save("random_forest_trees.gif", writer="pillow")
plt.show()
```
